# 实验三：基于 PyTorch 的岗位分类神经网络模型
本实验接续实验二，利用 KMeans 生成的聚类标签作为伪标签，使用 PyTorch 搭建 MLP 全连接神经网络进行模型训练。

```python
import os
import sys
import torch
import numpy as np
import matplotlib.pyplot as plt

PROJECT_ROOT = os.path.abspath('..')
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

from src.ml_engine.classifier import JobClassifierNN, train, predict
```

## 1. 准备数据
在实际生产中，我们会从前面保存好的 Numpy 缓存读取特征，这里我们模拟生成数据用于网络结构测试。

```python
# 模拟加载 Embedding 特征和伪标签
num_samples = 1000
embedding_dim = 1024
num_classes = 5

np.random.seed(42)
X_mock = np.random.randn(num_samples, embedding_dim).astype(np.float32)
y_mock = np.random.randint(0, num_classes, size=(num_samples,))

print(f"输入特征形状: {X_mock.shape}")
print(f"标签形状: {y_mock.shape}")
```

## 2. 定义与查看 PyTorch 网络结构

```python
model = JobClassifierNN(input_dim=embedding_dim, num_classes=num_classes)
print(model)

# 查看设备
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"当前计算设备: {device}")
```

## 3. 模型训练与评估

```python
print("开始训练...")
# 调用封装好的训练接口
test_acc, history, report = train(
    X=X_mock, 
    y=y_mock, 
    save_path=os.path.join(PROJECT_ROOT, 'models', 'mock_nn_classifier.pt'),
    epochs=10, 
    batch_size=32
)

print(f"\n最终测试集准确率: {test_acc:.4f}")
print("分类报告:\n", report)
```

## 4. 训练过程可视化

```python
plt.figure(figsize=(12, 5))

# Loss 曲线
plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='Train Loss', color='red')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Accuracy 曲线
plt.subplot(1, 2, 2)
plt.plot(history['val_acc'], label='Val Accuracy', color='blue')
plt.title('Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
