# 实验二：NLP文本向量化与 KMeans 聚类分析
本实验展示如何将招聘描述文本进行向量化处理，并利用 KMeans 算法进行聚类。

```python
import os
import sys
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

# 将项目根目录加入路径
PROJECT_ROOT = os.path.abspath('..')
if PROJECT_ROOT not in sys.path:
    sys.path.append(PROJECT_ROOT)

from src.data_pipeline.nlp_processor import clean_html_tags, tokenize, build_tfidf_matrix, build_embedding_matrix
```

## 1. 数据加载与基础预处理

```python
data_path = os.path.join(PROJECT_ROOT, 'data', 'cleaned_recruitment_data.csv')

# 假设文件已存在
try:
    df = pd.read_csv(data_path)
    print(f"加载了 {len(df)} 条数据")
    # 展示原始文本
    print(df['positionDetail'].head(2).tolist())
except Exception as e:
    print('数据加载失败:', e)
```

## 2. 大模型 Embedding 向量化
此处调用阿里云百炼 API（已配置限流和缓存），将文本转化为高维稠密特征。

```python
# 注意：需要先配置系统环境变量 DASHSCOPE_API_KEY
try:
    # 为了演示，仅抽样 50 条
    if 'df' in locals():
        embedding_matrix = build_embedding_matrix(df, sample_size=50)
        print(f"Embedding 特征矩阵形状: {embedding_matrix.shape}")
except Exception as e:
    print('Embedding提取失败:', e)
```

## 3. KMeans 聚类
使用 KMeans 对高维向量进行降维聚类，生成伪标签供后续神经网络使用。

```python
try:
    if 'embedding_matrix' in locals():
        num_clusters = 5
        kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
        labels = kmeans.fit_predict(embedding_matrix)
        print(f"聚类完成，前10个样本的分类标签: {labels[:10]}")
except Exception as e:
    print('聚类失败:', e)
```
